# 1. Data exploration

Given dataset have five variables: four independent ones and one dependent that needs to be predicted. Independent varibles represent the four sensor readings, while the dependent one represent vineyard yield.

```python
# Loading dataset from csv file.
vineyard_data = pd.read_csv("./data/podaci.csv", sep=",", names=["x1", "x2", "x3", "x4", "u"])
vineyard_data.head(5)
```

```python
# Getting to know the dataset.
print(f"{vineyard_data.isnull().sum()} \n")
print(f"{vineyard_data.dtypes} \n")
print(f"{vineyard_data.shape} \n")
```

```python
# Statistical measures for every variable.
vineyard_data.describe()
```

```python
# Pairplot will show us correlations between evey variable and its distribution.
sns.pairplot(data=vineyard_data)
```
